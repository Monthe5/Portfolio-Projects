```python
from typing import Counter
import fitz #this is from PyMuPDF also can be typed as from PyMuPDF import fitz to note which package is in use
import PIL.Image #from pillow
import io

pdf = fitz.open ("/Africas-Energy-Future-is-Renewables-Flagship2023.pdf")
counter = 1
for i in range(len(pdf)):
  page = pdf[i]
  images = page.get_images()
  for image in images:
    base_img = pdf.extract_image(image[0])
    image_data =  base_img["image"]
    img = PIL.Image.open(io.BytesIO(image_data))
    extension = base_img ['ext']
    img.save(open(f"image{counter}.{extension}", 'wb'))
    counter = +1
```
